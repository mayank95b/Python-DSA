## Accessing help docs of NumPy

```python
# Good documentation is available for numpy
import numpy as np
help(np.sort)
```

```python
# General-purpose documents like a glossary and help on the basic concepts
# of numpy are available under the ``doc`` sub-module:
from numpy import doc
help(doc)
```

```python
# To get the functions which contains the word "cos"
# <Shift> + <Enter>
np.*cos*?
```

```python
# To view the docsctring
np.cos?
```

```python
# To search for documents containing a keyword
# np.lookfor('keyword')
np.lookfor('sort')
```
